```python
def make_source_ascii(table):
    '''
    Inputs:
        table: A dataframe holding the information of a source or multiple sources, stacked in rows.
    Outputs:
        src: A string containing all of the information of the dataframe formatted as in ascii, split by line, appropriate for sedfitter fitting.
    '''
    src = ''
    for i in range(len(table)):
        src += " ".join(str(thingbob) for thingbob in table.iloc[i].to_list()) + '\n'
    return src
```

```python
df2 = pd.read_pickle('../data/raw/SESNA_INPUTS_JakeJune20.pkl')
df2.iloc[:,11:]
```

```python
# Normalization of SESNA on H band -- use deep copies to not end up changing the normalization factor each pass
norm_factor = df2['FNU_H'].copy()
df_norm = df2.copy()
# Normalize Fnu at each waveband
df_norm.iloc[:,11:26:2] = df_norm.iloc[:,11:26:2].div(norm_factor,axis=0)

# Because flux is a random variable, we need to multiply the error by the same factor as the flux
df_norm.iloc[:,12:27:2] = df_norm.iloc[:,12:27:2].div(norm_factor,axis=0)

df_norm

with open('../data/raw/SESNA_normalized','w') as output:
    output.write(make_source_ascii(df_norm))
```
