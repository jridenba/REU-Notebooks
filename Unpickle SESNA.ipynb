```python
def make_source_ascii(table):
    src = ''
    for i in range(len(table)):
        src += " ".join(str(thingbob) for thingbob in table.iloc[i].to_list()) + '\n'
    return src
```

```python
source_ascii = make_source_ascii(df)
print(source_ascii[0:20])
print(source_ascii[100:120])
print(source_ascii[200:220])
```

```python
df2 = pd.read_pickle('../data/raw/SESNA_INPUTS_JakeJune20.pkl')
df2.iloc[:,11:]
```

```python
df2.columns
```

```python
# Normalization of SESNA on H band, still need to propagate to errors
norm_factor = df2['FNU_H'].copy()
df_norm = df2.copy()
# Normalize Fnu at each waveband
df_norm.iloc[:,11:26:2] = df_norm.iloc[:,11:26:2].div(norm_factor,axis=0)

# Propagate this to the sigma Fnu's
sigB = df_norm['SIGMA_FNU_H'].div(norm_factor,axis=0)**2
sigA = df_norm.iloc[:,12:27:2].div(norm_factor,axis=0)**2
sig = np.sqrt(sigA.add(sigB,axis=0))
mag = df_norm.iloc[:,11:26:2].copy()
sig = sig.multiply(np.array(mag),axis=0)

df_norm.iloc[:,12:27:2] = sig
df_norm

with open('../data/raw/SESNA_normalized','w') as output:
    output.write(make_source_ascii(df_norm))
```
